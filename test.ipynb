```python
# À exécuter dans votre terminal si ce n'est pas fait :
# pip install soccerdata pandas

import soccerdata as sd
import pandas as pd

# Afficher toutes les colonnes pour une meilleure visibilité
pd.set_option('display.max_columns', None)
```

```python
# Initialisation du scraper FBref
# "FRA-Ligue 1" est l'identifiant standard pour le championnat de France
fbref = sd.FBref(leagues="FRA-Ligue 1", seasons="25-26")

print("API FBref initialisée pour la Ligue 1 (2025-2026)")
```

```python
# Lecture des stats par équipe
team_stats = fbref.read_team_season_stats(stat_type="standard")

# On affiche les 5 premières lignes
team_stats
```

```python
# Récupération du calendrier complet
schedule = fbref.read_schedule()

# Affichage des 5 derniers matchs joués
schedule[schedule['score'].notna()].tail(5)
```
