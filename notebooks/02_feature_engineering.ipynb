# 02 - Feature Engineering

Objectif: construire le dataset match-level et inspecter les features.

```python
from pathlib import Path
import pandas as pd
from build_dataset import build_dataset

DATA_DIR = Path("data")
RAW_DIR = DATA_DIR / "raw"
PROC_DIR = DATA_DIR / "processed"

intl_path = RAW_DIR / "international_results.csv"
rank_path = RAW_DIR / "fifa_rankings.csv"
squad_path = RAW_DIR / "squads.csv"
player_stats_path = RAW_DIR / "fbref" / "player_standard.csv"

dataset_out = PROC_DIR / "match_dataset.csv"
team_features_out = PROC_DIR / "national_team_features.csv"
elo_out = PROC_DIR / "elo_ratings.csv"
form_out = PROC_DIR / "matches_with_form.csv"

if intl_path.exists():
    build_dataset(
        international_results=intl_path,
        fifa_rankings=rank_path if rank_path.exists() else None,
        squad_csv=squad_path if squad_path.exists() else None,
        player_stats=player_stats_path if player_stats_path.exists() else None,
        dataset_out=dataset_out,
        team_features_out=team_features_out,
        elo_out=elo_out,
        form_out=form_out,
    )
else:
    print("Missing international_results.csv in data/raw")
```

```python
if dataset_out.exists():
    df = pd.read_csv(dataset_out)
    display(df.head())
    print(df.shape)
    print("Feature columns sample:")
    print(df.columns[:25])
    print("Missing rate top 10:")
    print(df.isna().mean().sort_values(ascending=False).head(10))
```

```python
from pathlib import Path

OUTPUT_DIR = Path("outputs")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

if dataset_out.exists():
    df = pd.read_csv(dataset_out)
    df.head(50).to_csv(OUTPUT_DIR / "feature_preview.csv", index=False)
    missing = df.isna().mean().sort_values(ascending=False)
    missing.to_csv(OUTPUT_DIR / "feature_missing_rate.csv")
    pd.Series(df.columns, name="feature").to_csv(OUTPUT_DIR / "feature_list.csv", index=False)
```
