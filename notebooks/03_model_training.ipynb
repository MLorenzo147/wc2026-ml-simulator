# 03 - Model Training

Objectif: entrainer champion + challengers et sauvegarder les artefacts.

```python
from pathlib import Path
import subprocess

dataset_path = Path("data/processed/match_dataset.csv")
models_dir = Path("data/processed/models")

if dataset_path.exists():
    cmd = ["python", "train_models.py", "--dataset", str(dataset_path), "--models-dir", str(models_dir)]
    print("Running:", " ".join(cmd))
    subprocess.run(cmd, check=False)
else:
    print("Missing match_dataset.csv. Run feature engineering first.")
```

```python
import json

feature_path = models_dir / "feature_columns.json"
label_path = models_dir / "label_classes.json"

if feature_path.exists():
    feature_cols = json.loads(feature_path.read_text())
    print("Feature count:", len(feature_cols))

if label_path.exists():
    labels = json.loads(label_path.read_text())
    print("Label classes:", labels)
```

```python
import json
from pathlib import Path

OUTPUT_DIR = Path("outputs")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

artifacts = {
    "models_dir": str(models_dir),
    "feature_columns_path": str(feature_path),
    "label_classes_path": str(label_path)
}
(OUTPUT_DIR / "training_artifacts.json").write_text(json.dumps(artifacts, indent=2))
```
