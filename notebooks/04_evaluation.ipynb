# 04 - Evaluation

Objectif: comparer les modeles et produire les metriques.

```python
from pathlib import Path
import subprocess
import pandas as pd

dataset_path = Path("data/processed/match_dataset.csv")
models_dir = Path("data/processed/models")
metrics_path = Path("data/processed/metrics.csv")

if dataset_path.exists() and models_dir.exists():
    cmd = ["python", "evaluate.py", "--dataset", str(dataset_path), "--models-dir", str(models_dir), "--output", str(metrics_path)]
    print("Running:", " ".join(cmd))
    subprocess.run(cmd, check=False)
else:
    print("Missing dataset or models. Run training first.")

if metrics_path.exists():
    metrics_df = pd.read_csv(metrics_path)
    display(metrics_df.sort_values("log_loss"))
```

```python
from pathlib import Path

OUTPUT_DIR = Path("outputs")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

if metrics_path.exists():
    metrics_df = pd.read_csv(metrics_path)
    metrics_df.to_csv(OUTPUT_DIR / "model_metrics.csv", index=False)
```
