# 05 - WC2026 Simulation

Objectif: simuler la Coupe du Monde 2026 avec un modele calibre.

```python
from pathlib import Path
import subprocess
import pandas as pd

model_path = Path("data/processed/models/champion_xgboost.pkl")
models_dir = Path("data/processed/models")
feature_cols = models_dir / "feature_columns.json"
label_classes = models_dir / "label_classes.json"

intl_path = Path("data/raw/international_results.csv")
rank_path = Path("data/raw/fifa_rankings.csv")
team_features = Path("data/processed/national_team_features.csv")
groups_path = Path("data/raw/wc2026_groups.csv")
schedule_path = Path("data/raw/wc2026_schedule.csv")
output_path = Path("data/processed/wc2026_simulation.csv")

ready = all([
    model_path.exists(),
    feature_cols.exists(),
    label_classes.exists(),
    intl_path.exists(),
    groups_path.exists(),
    schedule_path.exists(),
])

if ready:
    cmd = [
        "python", "simulate_wc2026.py",
        "--model-path", str(model_path),
        "--feature-columns", str(feature_cols),
        "--label-classes", str(label_classes),
        "--historical-matches", str(intl_path),
        "--fifa-rankings", str(rank_path),
        "--team-features", str(team_features),
        "--groups-csv", str(groups_path),
        "--schedule-csv", str(schedule_path),
        "--output", str(output_path),
        "--n-simulations", "1000",
    ]
    print("Running:", " ".join(cmd))
    subprocess.run(cmd, check=False)
else:
    print("Missing files for simulation. Check raw data and trained model.")

if output_path.exists():
    results_df = pd.read_csv(output_path)
    display(results_df.head(20))
```

```python
from pathlib import Path

OUTPUT_DIR = Path("outputs")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

if output_path.exists():
    results_df = pd.read_csv(output_path)
    results_df.head(20).to_csv(OUTPUT_DIR / "wc2026_simulation_top20.csv", index=False)
```
